```python
import pandas as pd
from collections import Counter

def find_most_frequent_main_numbers(df, n=None):
    """
    Finds the most frequent numbers in each of the main ball columns
    (Ball 1 to Ball 5) of a EuroMillions draw history DataFrame.

    Args:
        df (pd.DataFrame): The DataFrame containing the draw data with
                           columns 'Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5'.
        n (int, optional): The number of most frequent numbers to return for
                           each ball column. If None, returns all frequencies.
                           Defaults to None.

    Returns:
        dict: A dictionary where keys are the ball column names and values
              are Counter objects showing the frequency of numbers in that column.
    """
    if not all(col in df.columns for col in ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']):
        print("Error: DataFrame must contain columns 'Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5'.")
        return {}

    frequent_numbers = {}
    ball_columns = ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']

    for col in ball_columns:
        frequent_numbers[col] = Counter(df[col])
        print(f"\nMost Frequent Numbers in {col}:")
        if n is None:
            for number, count in frequent_numbers[col].most_common():
                print(f"  Number {number}: {count} times")
        else:
            for number, count in frequent_numbers[col].most_common(n):
                print(f"  Number {number}: {count} times")

    return frequent_numbers

if __name__ == "__main__":
    # Load your EuroMillions draw history data
    df = pd.read_csv('/content/euromillions-draw-history.csv')

    # Find and print all frequent numbers for each main ball column
    all_frequent_main = find_most_frequent_main_numbers(df)

    # Find and print the top 10 most frequent numbers for each main ball column
    print("\n--- Top 10 Most Frequent Numbers in Each Main Ball Column ---")
    top_10_frequent_main = find_most_frequent_main_numbers(df, n=10)
```

```python
def find_most_frequent_main_numbers(df, n=5):
    """
    Finds the top N most frequent numbers in each of the main ball columns
    (Ball 1 to Ball 5) of a EuroMillions draw history DataFrame.

    Args:
        df (pd.DataFrame): The DataFrame containing the draw data with
                           columns 'Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5'.
        n (int): The number of most frequent numbers to return for
                 each ball column. Defaults to 5.

    Returns:
        dict: A dictionary where keys are the ball column names and values
              are lists of the top N (number, count) tuples.
    """
    if not all(col in df.columns for col in ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']):
        print("Error: DataFrame must contain columns 'Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5'.")
        return {}

    top_frequent = {}
    ball_columns = ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']

    for col in ball_columns:
        frequent_numbers = Counter(df[col])
        top_frequent[col] = frequent_numbers.most_common(n)
        print(f"\nTop {n} Most Frequent Numbers in {col}:")
        for number, count in top_frequent[col]:
            print(f"  Number {number}: {count} times")

    return top_frequent

if __name__ == "__main__":
    # Load your EuroMillions draw history data
    df = pd.read_csv('/content/euromillions-draw-history.csv')

    # Find and print the top 5 most frequent numbers for each main ball column
    top_5_frequent_main = find_most_frequent_main_numbers(df, n=5)
```

```python
import pandas as pd
from collections import Counter

def find_top_n_most_frequent_all_balls(df, n=5):
    """
    Finds the top N most frequent numbers across all main ball columns
    (Ball 1 to Ball 5) of a EuroMillions draw history DataFrame.

    Args:
        df (pd.DataFrame): The DataFrame containing the draw data with
                           columns 'Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5'.
        n (int): The number of most frequent numbers to return. Defaults to 5.

    Returns:
        list: A list of the top N (number, total_count) tuples, sorted by frequency.
    """
    if not all(col in df.columns for col in ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']):
        print("Error: DataFrame must contain columns 'Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5'.")
        return []

    all_balls_numbers = []
    ball_columns = ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']

    for col in ball_columns:
        all_balls_numbers.extend(df[col].tolist())

    all_balls_counts = Counter(all_balls_numbers)
    top_n_frequent = all_balls_counts.most_common(n)

    print(f"Top {n} Most Frequent Numbers Across All Main Balls:")
    for number, count in top_n_frequent:
        print(f"  Number {number}: {count} times")

    return top_n_frequent

if __name__ == "__main__":
    # Load your EuroMillions draw history data
    df = pd.read_csv('/content/euromillions-draw-history.csv')

    # Find and print the top 5 most frequent numbers across all main balls
    top_5_frequent_all = find_top_n_most_frequent_all_balls(df, n=5)
```

```python
import pandas as pd

def find_rows_with_at_least_n_matches(df, target_numbers, n=3):
    """
    Finds rows in a EuroMillions draw history DataFrame where at least N
    of the target numbers appear in the main ball columns (Ball 1 to Ball 5).

    Args:
        df (pd.DataFrame): The DataFrame containing the draw data with
                           columns 'Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5'.
        target_numbers (list): A list of numbers to check for in each row.
        n (int): The minimum number of target numbers that must appear in a row.
                 Defaults to 3.

    Returns:
        pd.DataFrame: A DataFrame containing the rows that meet the criteria.
                      Returns an empty DataFrame if no rows match.
    """
    if not all(col in df.columns for col in ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']):
        print("Error: DataFrame must contain columns 'Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5'.")
        return pd.DataFrame()

    matching_rows = []
    ball_columns = ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']

    for index, row in df.iterrows():
        drawn_numbers = row[ball_columns].tolist()
        matches = 0
        for number in target_numbers:
            if number in drawn_numbers:
                matches += 1
        if matches >= n:
            matching_rows.append(row)

    return pd.DataFrame(matching_rows)

if __name__ == "__main__":
    # Load your EuroMillions draw history data
    df = pd.read_csv('/content/euromillions-draw-history.csv')

    # The top 5 most frequent numbers you provided
    top_5_frequent = [29, 10, 33, 42, 50]

    # Find rows where at least 3 of these numbers appear
    matching_df = find_rows_with_at_least_n_matches(df, top_5_frequent, n=3)

    if not matching_df.empty:
        print(f"Rows where at least 3 of the numbers {top_5_frequent} appeared:")
        print(matching_df)
    else:
        print(f"No rows found where at least 3 of the numbers {top_5_frequent} appeared.")
```

```python
import pandas as pd

def find_rows_with_ball_and_ls(df, target_balls, ls1_targets, ls2_targets, min_ball_matches=3):
    """
    Finds rows where at least a certain number of target balls appear AND
    Lucky Star 1 matches one of the LS1 targets OR Lucky Star 2 matches one
    of the LS2 targets.

    Args:
        df (pd.DataFrame): The DataFrame containing the draw data.
        target_balls (list): A list of target main ball numbers.
        ls1_targets (list): A list of target Lucky Star 1 numbers.
        ls2_targets (list): A list of target Lucky Star 2 numbers.
        min_ball_matches (int): The minimum number of target balls that must appear.
                                Defaults to 3.

    Returns:
        pd.DataFrame: A DataFrame containing the matching rows.
    """
    if not all(col in df.columns for col in ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5', 'Lucky Star 1', 'Lucky Star 2']):
        print("Error: DataFrame must contain all required columns.")
        return pd.DataFrame()

    matching_rows = []
    ball_columns = ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']

    for index, row in df.iterrows():
        drawn_balls = row[ball_columns].tolist()
        drawn_ls1 = row['Lucky Star 1']
        drawn_ls2 = row['Lucky Star 2']

        ball_matches = 0
        for ball in target_balls:
            if ball in drawn_balls:
                ball_matches += 1

        ls1_match = drawn_ls1 in ls1_targets
        ls2_match = drawn_ls2 in ls2_targets

        if ball_matches >= min_ball_matches and (ls1_match or ls2_match):
            matching_rows.append(row)

    return pd.DataFrame(matching_rows)

if __name__ == "__main__":
    # Load your EuroMillions draw history data
    df = pd.read_csv('/content/euromillions-draw-history.csv')

    # Target main ball numbers
    target_main_balls = [29, 10, 33, 42, 50]
    min_matches = 3

    # Target Lucky Star numbers
    target_ls1 = [1, 2, 5, 6]
    target_ls2 = [9, 12, 7, 8, 10, 11]

    # Find rows that meet the criteria
    matching_df = find_rows_with_ball_and_ls(df, target_main_balls, target_ls1, target_ls2, min_matches)

    if not matching_df.empty:
        print(f"Rows with at least {min_matches} of the balls {target_main_balls} and")
        print(f"Lucky Star 1 in {target_ls1} OR Lucky Star 2 in {target_ls2}:")
        print(matching_df)
    else:
        print(f"No rows found with at least {min_matches} of the balls {target_main_balls} and")
        print(f"Lucky Star 1 in {target_ls1} OR Lucky Star 2 in {target_ls2}.")
```

```python
from math import comb

def probability_event():
    """
    Calculates the probability of the specified EuroMillions event occurring
    in a single future draw.

    Event:
    - At least 3 of the main balls are from the set [29, 10, 33, 42, 50].
    - AND (Lucky Star 1 is one of [1, 2, 5, 6] OR Lucky Star 2 is one of [9, 12, 7, 8, 10, 11]).
    """
    # Probability of at least 3 target main balls
    total_main_combinations = comb(50, 5)
    target_balls = 5
    non_target_balls = 45

    prob_3_matches = (comb(target_balls, 3) * comb(non_target_balls, 2)) / total_main_combinations
    prob_4_matches = (comb(target_balls, 4) * comb(non_target_balls, 1)) / total_main_combinations
    prob_5_matches = (comb(target_balls, 5) * comb(non_target_balls, 0)) / total_main_combinations

    prob_at_least_3_target_balls = prob_3_matches + prob_4_matches + prob_5_matches

    # Probability of the Lucky Star condition
    ls1_target = 4
    ls2_target = 6
    total_ls = 12

    prob_ls1_match = ls1_target / total_ls
    prob_ls2_match = ls2_target / total_ls
    prob_both_match = prob_ls1_match * prob_ls2_match  # Assuming independence

    prob_ls1_or_ls2_match = prob_ls1_match + prob_ls2_match - prob_both_match

    # Probability of both events occurring (assuming independence)
    probability = prob_at_least_3_target_balls * prob_ls1_or_ls2_match

    return probability

if __name__ == "__main__":
    probability = probability_event()
    print(f"The probability of the specified event in a single EuroMillions draw is approximately: {probability:.8f}")
```

```python
from math import comb

def probability_event():
    """
    Calculates the probability of the specified EuroMillions event occurring
    in a single future draw.

    Event:
    - At least 3 of the main balls are from the set [29, 10, 33, 42, 50].
    - AND (Lucky Star 1 is one of [1, 2, 5, 6] OR Lucky Star 2 is one of [9, 12, 7, 8, 10, 11]).
    """
    # Probability of at least 3 target main balls
    total_main_combinations = comb(50, 5)
    target_balls = 5
    non_target_balls = 45

    prob_3_matches = (comb(target_balls, 3) * comb(non_target_balls, 2)) / total_main_combinations
    prob_4_matches = (comb(target_balls, 4) * comb(non_target_balls, 1)) / total_main_combinations
    prob_5_matches = (comb(target_balls, 5) * comb(non_target_balls, 0)) / total_main_combinations

    prob_at_least_3_target_balls = prob_3_matches + prob_4_matches + prob_5_matches

    # Probability of the Lucky Star condition
    ls1_target = 4
    ls2_target = 6
    total_ls = 12

    prob_ls1_match = ls1_target / total_ls
    prob_ls2_match = ls2_target / total_ls
    prob_both_match = prob_ls1_match * prob_ls2_match  # Assuming independence

    prob_ls1_or_ls2_match = prob_ls1_match + prob_ls2_match - prob_both_match

    # Probability of both events occurring (assuming independence)
    probability = prob_at_least_3_target_balls * prob_ls1_or_ls2_match

    return probability

def calculate_expected_cost_per_event(probability_per_draw, num_draws, cost_per_ticket):
    """
    Calculates the expected cost associated with the occurrence of a specific
    event over a number of draws. This is a theoretical average.

    Args:
        probability_per_draw (float): The probability of the event in a single draw.
        num_draws (int): The number of draws considered.
        cost_per_ticket (float): The cost of one EuroMillions ticket.

    Returns:
        float: The expected cost per occurrence of the event. Returns infinity
               if the expected number of occurrences is zero.
    """
    expected_occurrences = probability_per_draw * num_draws
    if expected_occurrences > 0:
        total_cost = cost_per_ticket * num_draws
        expected_cost_per_event = total_cost / expected_occurrences
        return expected_cost_per_event
    else:
        return float('inf')

if __name__ == "__main__":
    probability_single_draw = probability_event()
    num_draws = 100
    cost_per_ticket = 2.50  # Price in Ireland

    expected_cost = calculate_expected_cost_per_event(probability_single_draw, num_draws, cost_per_ticket)

    print(f"The probability of the specified event in a single EuroMillions draw is approximately: {probability_single_draw:.8f}")
    print(f"Considering {num_draws} draws with a ticket price of €{cost_per_ticket:.2f}:")
    if expected_cost == float('inf'):
        print("The expected number of occurrences of this event is zero over 100 draws.")
    else:
        print(f"The expected cost associated with each occurrence of this event is approximately: €{expected_cost:.2f}")

    print("\nImportant Note:")
    print("This is a theoretical calculation based on probability.")
    print("The lottery is random, and actual outcomes can vary significantly.")
    print("This does not guarantee that you will spend this exact amount to see this event occur.")
    print("Play responsibly.")
```

```python
from collections import Counter
from itertools import combinations

def get_top_frequent_numbers_per_ball(df, n=5):
    """
    Finds the top N most frequent numbers in each of the main ball columns.

    Args:
        df (pd.DataFrame): The DataFrame containing the draw data.
        n (int): The number of top frequent numbers to return.

    Returns:
        dict: A dictionary where keys are ball column names and values are
              lists of the top N most frequent numbers.
    """
    top_frequent = {}
    ball_columns = ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']
    for col in ball_columns:
        counts = Counter(df[col])
        top_n = [num for num, count in counts.most_common(n)]
        top_frequent[col] = top_n
    return top_frequent

def generate_potential_combinations(top_frequent):
    """
    Generates a few potential 5-number combinations based on the top
    frequent numbers from each ball position. This is NOT a statistically
    sound prediction method.

    Args:
        top_frequent (dict): A dictionary of top frequent numbers per ball.

    Returns:
        list: A list of 5 potential main number combinations.
    """
    combinations_list = []

    # Try combining the top number from each ball (if unique)
    top_from_each = []
    seen = set()
    for col in ['Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5']:
        if top_frequent[col] and top_frequent[col][0] not in seen:
            top_from_each.append(top_frequent[col][0])
            seen.add(top_frequent[col][0])

    if len(top_from_each) >= 5:
        combinations_list.append(tuple(sorted(top_from_each[:5])))

    # Try other combinations using the top 2-3 from each ball (avoiding repeats)
    all_top = []
    for col in top_frequent:
        all_top.extend(top_frequent[col][:3])
    unique_top = sorted(list(set(all_top)))

    # Generate a few combinations (this is a very basic approach)
    for combo in combinations(unique_top, 5):
        if len(combinations_list) < 5:
            combinations_list.append(combo)
        else:
            break

    return combinations_list

if __name__ == "__main__":
    import pandas as pd

    # Load your EuroMillions draw history data
    try:
        df = pd.read_csv('/content/euromillions-draw-history.csv')
    except FileNotFoundError:
        print("Error: EuroMillions draw history CSV file not found at /content/")
        exit()

    top_frequent_numbers = get_top_frequent_numbers_per_ball(df)
    print("Top 5 Most Frequent Numbers in Each Ball Position:")
    for ball, numbers in top_frequent_numbers.items():
        print(f"  {ball}: {numbers}")

    potential_main_combinations = generate_potential_combinations(top_frequent_numbers)

    lucky_star_1_targets = [1, 2, 5, 6]
    lucky_star_2_targets = [9, 12, 7, 8, 10, 11]

    print("\n5 Potential Combinations Based on Frequent Numbers (NOT a prediction):")
    for i, combo in enumerate(potential_main_combinations):
        print(f"  Combination {i+1}: {sorted(list(combo))}")

    print("\nPossible Combinations for Prize Tiers (Based on Frequent Numbers and Lucky Star Sets):")

    print("\n5 + 1 (5 Main Numbers + 1 Lucky Star):")
    for combo in potential_main_combinations:
        for ls1 in lucky_star_1_targets[:min(len(lucky_star_1_targets), 2)]:
            print(f"  {sorted(list(combo))} + LS1: {ls1}")

    print("\n5 (5 Main Numbers):")
    for combo in potential_main_combinations:
        print(f"  {sorted(list(combo))} + (Any two Lucky Stars)") # Lucky Stars don't matter for this tier

    print("\n5 + 2 (5 Main Numbers + 2 Lucky Stars):")
    for combo in potential_main_combinations:
        for ls1 in lucky_star_1_targets[:min(len(lucky_star_1_targets), 2)]:
            for ls2 in lucky_star_2_targets[:min(len(lucky_star_2_targets), 2)]:
                if ls1 != ls2:
                    print(f"  {sorted(list(combo))} + LS1: {ls1}, LS2: {ls2}")

    print("\nImportant Note:")
    print("This code generates combinations based on historical frequency, which is NOT a reliable predictor of future lottery outcomes.")
    print("EuroMillions draws are random, and every number has an equal chance of being drawn in each draw.")
    print("Play responsibly and understand that lottery results are based on chance.")
```

```python
import pandas as pd

def check_combinations_in_draws(df, combinations_to_check):
    """
    Checks if specific main number and Lucky Star combinations are present
    in the draw history.

    Args:
        df (pd.DataFrame): The DataFrame containing the draw data.
        combinations_to_check (list): A list of tuples, where each tuple is:
                                      (main_numbers_list, lucky_star_1, lucky_star_2).

    Returns:
        dict: A dictionary where keys are the combinations and values are boolean
              (True if found, False otherwise).
    """
    results = {}
    for main_numbers, ls1, ls2 in combinations_to_check:
        found = False
        sorted_main = sorted(main_numbers)
        for _, row in df.iterrows():
            draw_main = sorted([row['Ball 1'], row['Ball 2'], row['Ball 3'], row['Ball 4'], row['Ball 5']])
            draw_ls1 = row['Lucky Star 1']
            draw_ls2 = row['Lucky Star 2']
            if draw_main == sorted_main and draw_ls1 == ls1 and draw_ls2 == ls2:
                found = True
                break
        results[(tuple(main_numbers), ls1, ls2)] = found
    return results

if __name__ == "__main__":
    # Load your EuroMillions draw history data
    try:
        df = pd.read_csv('/content/euromillions-draw-history.csv')
    except FileNotFoundError:
        print("Error: EuroMillions draw history CSV file not found at /content/")
        exit()

    combinations_to_check = [
        ([1, 14, 29, 42, 50], 1, 9),
        ([1, 14, 29, 42, 50], 1, 12),
        ([1, 14, 29, 42, 50], 2, 9),
        ([1, 14, 29, 42, 50], 2, 12),
        ([1, 4, 10, 11, 14], 1, 9),
        ([1, 4, 10, 11, 14], 1, 12),
        ([1, 4, 10, 11, 14], 2, 9),
        ([1, 4, 10, 11, 14], 2, 12),
        ([1, 4, 10, 11, 15], 1, 9),
        ([1, 4, 10, 11, 15], 1, 12),
        ([1, 4, 10, 11, 15], 2, 9),
        ([1, 4, 10, 11, 15], 2, 12),
        ([1, 4, 10, 11, 19], 1, 9),
        ([1, 4, 10, 11, 19], 1, 12),
        ([1, 4, 10, 11, 19], 2, 9),
        ([1, 4, 10, 11, 19], 2, 12),
        ([1, 4, 10, 11, 25], 1, 9),
        ([1, 4, 10, 11, 25], 1, 12),
        ([1, 4, 10, 11, 25], 2, 9),
        ([1, 4, 10, 11, 25], 2, 12),
    ]

    results = check_combinations_in_draws(df, combinations_to_check)

    print("Checking for the presence of specific combinations in the draw history:")
    for combo, present in results.items():
        main_numbers, ls1, ls2 = combo
        print(f"Main Numbers: {sorted(list(main_numbers))}, LS1: {ls1}, LS2: {ls2} - Present: {present}")
```
