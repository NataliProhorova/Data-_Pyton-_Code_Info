```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Standardize the numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[numerical_features])


# Perform PCA PCA  transforms the data into a new set of features called principal components (PCs).  These PCs capture the most significant variations in the original data.
pca = PCA()
pca.fit(X_scaled)
```

```python
#print a list of values representing the proportion of variance explained by each principal component.
explained_variance_ratio = pca.explained_variance_ratio_
print(explained_variance_ratio)
```

```python
# calculates and prints the total number of principal components generated by PCA.
n_components = len(explained_variance_ratio)
print(f"Number of Components: {n_components}")
```

```python
principal_components = pca.components_
print(principal_components.shape)  # Prints the dimensions of the components matrix
```

```python
# Apply PCA to reduce dimensionality
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)



# Create a new DataFrame with the reduced features
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(n_components)])

df_pca.head()
```
